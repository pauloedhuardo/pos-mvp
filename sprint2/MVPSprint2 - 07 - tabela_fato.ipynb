### Construindo a tabela fato_cumprimento_mandado

```
%sql
SELECT * FROM silver.cumprimento AS C
  INNER JOIN silver.mandado AS M ON M.mdid = C.cpcodmandado
  INNER JOIN silver.tipo_mandado AS T ON M.mdcodtpmandado = T.tmid
  INNER JOIN silver.oficial_justica AS O ON M.mdcodoficialjustica = O.ojid
LIMIT 1000
```

```
%sql
DROP TABLE IF EXISTS gold.fato_cumprimento_mandado;
```

```
%sql
CREATE TABLE gold.fato_cumprimento_mandado AS
SELECT C.cpid AS codigo_cumprimento, 
       C.cporigem AS origem_cumprimento, 
       M.mdid AS codigo_mandado, 
       T.tmdescricao AS tipo_mandado, 
       O.ojnome AS nome_ofical_justica 
FROM silver.cumprimento AS C
  INNER JOIN silver.mandado AS M ON M.mdid = C.cpcodmandado
  INNER JOIN silver.tipo_mandado AS T ON M.mdcodtpmandado = T.tmid
  INNER JOIN silver.oficial_justica AS O ON M.mdcodoficialjustica = O.ojid
```

```
%sql
SELECT * FROM gold.fato_cumprimento_mandado LIMIT 1000;
```
