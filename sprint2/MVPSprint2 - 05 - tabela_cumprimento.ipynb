Seleciona somente os registros que de fato tenham o cumprimento finalizado e transforma os dados da coluna "cporigem".

```
%sql
DROP TABLE silver.cumprimento
```

```
%sql
CREATE TABLE silver.cumprimento AS
SELECT C.cpid, C.cpdtfato, C.cpfinalizado, C.cpcodmandado,
CASE
    WHEN C.cporigem = 0 THEN "APLICATIVO"
    WHEN C.cporigem = 1 THEN "SISTEMA_JUDICIAL"
END AS cporigem
FROM bronze.cumprimento AS C
WHERE C.cpfinalizado IS TRUE;
```
